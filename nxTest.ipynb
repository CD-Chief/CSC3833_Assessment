```python
import networkx as nx
import altair as alt
import nx_altair as nxa

# Define country neighbors as a dictionary
country_borders = {
    'BRA': ['ARG', 'BOL'],
    'ARG': ['BRA', 'BOL'],
    'BOL': ['BRA', 'ARG']
}

# Build the graph
G = nx.Graph()
for country, neighbors in country_borders.items():
    for neighbor in neighbors:
        G.add_edge(country, neighbor)

# Get a layout for node positions (spring layout for clarity)
pos = nx.spring_layout(G, seed=42)  # You can use latitude/longitude if available

# Draw the network graph using nx_altair
viz = nxa.draw_networkx(G, pos)
viz = viz.interactive()

# Show in Jupyter or export to HTML
viz.show()
```
